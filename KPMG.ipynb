```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
```

```python
df = pd.read_excel("Dataaa.csv")
df.head()
```

```python
mass_cust = df[df['wealth_segment'] == 'Mass Customer']
```

```python
df.drop(["customer_id", "first_name", "last_name", ], axis = 1, inplace = True)
df.head(10)
```

```python
df = df.dropna()
```

```python
le1 = LabelEncoder()
df['gender'] = le1.fit_transform(df['gender'])

le2 = LabelEncoder()
df['job_title'] = le2.fit_transform(df['job_title'])

le3 = LabelEncoder()
df['job_industry_category'] = le3.fit_transform(df['job_industry_category'])

le4 = LabelEncoder()
df['wealth_segment'] = le4.fit_transform(df['wealth_segment'])

le5 = LabelEncoder()
df['deceased_indicator'] = le5.fit_transform(df['deceased_indicator'])

le6 = LabelEncoder()
df['owns_car'] = le6.fit_transform(df['owns_car'])

df.head()
```

```python
df.head(10)
```

```python
df.shape
```

```python
k_range = range(1, 11)
score = []
for i in k_range:
    model = KMeans(n_clusters = i)
    model.fit_predict(df)
    score.append(model.inertia_)
```

```python
plt.plot(k_range, score)
plt.plot()
```

```python
real_model = KMeans(3)
y_pred = real_model.fit_predict(df)
df['groups'] = y_pred
```

```python
df.head(10)
```

```python
group0 = df[df['groups'] == 0]
group1 = df[df['groups'] == 1]
group2 = df[df['groups'] == 2]

print(group0['past_3_years_bike_related_purchases'].mean())
print(group1['past_3_years_bike_related_purchases'].mean())
print(group2['past_3_years_bike_related_purchases'].mean())
```
